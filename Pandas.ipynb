```python
import pandas as pd

data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/')
#data_pd = pd.DataFrame(data, columns=['POS','TYTUŁ','ARTYSTA','ROK','MAX POZ'])
data_pd = pd.DataFrame(data[0])
data_pd.columns = data_pd.iloc[0]
data_pd = data_pd.iloc[1:]
data_pd=data_pd.rename(columns={'TITLE':'Tytul', 'ARTIST':'Artysta','YEAR':'Rok','HIGH POSN':'MAX POZ'})
data_pd['Artysta'].nunique()

             
```

```python
df_min_year = data_pd.copy()

df_min_year['Rok_nm'] = pd.to_numeric(df_min_year['Rok'])


df_min_year = df_min_year.loc[df_min_year.groupby('Artysta')['Rok_nm'].idxmin()]

df_min_year[['Artysta', 'Rok']].sort_values(by='Artysta')

df_min_year.to_csv('raw_data.csv', index=False)

```
